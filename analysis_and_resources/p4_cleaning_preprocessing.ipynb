```python
import pandas as pd
```

```python
file_path = 'data/shopping.csv' 
data = pd.read_csv(file_path)
data.head()
```

```python
data.info()
```

```python
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns
```

```python
# Fill missing values in numerical columns with mean
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].mean())
```

```python
# Fill missing values in categorical columns with mode
for col in categorical_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data['Transaction_Date'] = pd.to_datetime(data['Transaction_Date'], errors='coerce')
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data.head()
```

```python
data.drop_duplicates(inplace=True)
data.drop(columns=['Date'], inplace=True)
```

```python
data.rename(columns={
    'CustomerID': 'Customer_ID',
    'Transaction_ID': 'Transaction_ID',
    'Product_SKU': 'Product_SKU',
    'Product_Description': 'Product_Description',
    'Product_Category': 'Product_Category',
    'Avg_Price': 'Average_Price',
    'Coupon_Status': 'Coupon_Status',
    'Offline_Spend': 'Offline_Spend',
    'Online_Spend': 'Online_Spend',
    'Coupon_Code': 'Coupon_Code',
    'Discount_pct': 'Discount_Percentage'
}, inplace=True)
data.head()
```

```python
# Determine categorical columns excluding datetime types
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

# Convert categorical columns to category type for memory optimization
for col in categorical_cols:
    data[col] = data[col].astype('category')
```

```python
#Add a column to indicate if the transaction date is close to or on a holiday
# List of holidays in 2019
holidays_2019 = {
    '2019-01-01': ('New Year\'s Day', 7),  
    '2019-01-21': ('Martin Luther King Jr. Day', 0),
    '2019-02-18': ('Washington\'s Birthday', 0),
    '2019-05-27': ('Memorial Day', 5),
    '2019-06-19': ('Juneteenth', 5),
    '2019-07-04': ('Independence Day', 5),
    '2019-09-02': ('Labor Day', 5),
    '2019-10-14': ('Columbus Day', 0),
    '2019-11-11': ('Veterans Day', 5),
    '2019-11-28': ('Thanksgiving Day', 14),
    '2019-12-25': ('Christmas Day', 14),  
    '2019-07-15': ('Amazon Prime Day', 0),  
    '2019-07-16': ('Amazon Prime Day', 0),
    '2019-11-29': ('Black Friday', 0),  
    '2019-11-30': ('Small Business Saturday', 0),  
    '2019-12-02': ('Cyber Monday', 0),  
    '2019-12-09': ('Green Monday', 0)
}
holidays_2019 = {pd.to_datetime(date): (name, pd.Timedelta(days=days)) for date, (name, days) in holidays_2019.items()}

```

```python
# Create a new column to indicate which holiday the transaction date is close to
def find_nearest_holiday_custom(date):
    if pd.isnull(date):
        return None
    for holiday_date, (holiday_name, threshold) in holidays_2019.items():
        if abs(date - holiday_date) <= threshold:
            return holiday_name
    return None

# Apply the function to create the 'Near_Holiday' column
data['Near_Holiday'] = data['Transaction_Date'].apply(find_nearest_holiday_custom)

# Display the updated DataFrame with the new column
data.head()
```

```python
# Display all unique values in the 'Location' column
unique_locations = data['Location'].unique()

# Print each unique location
for location in unique_locations:
    print(location)
```

```python
# Replace 'Chicago' with 'Illinois' in the 'Location' column
data['Location'] = data['Location'].replace({'Chicago': 'Illinois'})

# If you want to update the 'State' column after this change:
data['State'] = data['Location']

# Drop the original 'Location' column if no longer needed
data.drop(columns=['Location'], inplace=True)

# Display the updated DataFrame
data.head()
```

```python
# Display all unique values in the 'Product_Category' column
unique_categories = data['Product_Category'].unique()

# Print each unique category
for category in unique_categories:
    print(category)
```

```python
# Display all unique values in the 'Product_Description' column
unique_descriptions = data['Product_Description'].unique()

# Print each unique description
for description in unique_descriptions:
    print(description)
```

```python
# Define mappings for 'Product_Category' replacements
category_mappings = {
    'Nest-USA': 'Nest Smart Devices',
    'Nest-Canada': 'Nest Smart Devices',
    'Nest': 'Nest Smart Devices',
    'Google': 'Google Products',
    'Office': 'Office Supplies',
    'Notebooks & Journals': 'Stationery',
    'Notebooks': 'Stationery',
    'Bags': 'Bags & Backpacks',
    'More Bags': 'Bags & Backpacks',
    'Backpacks': 'Bags & Backpacks',
    'Gift Cards': 'Gift Cards',
    'Drinkware': 'Drinkware & Bottles',
    'Bottles': 'Drinkware & Bottles',
    'Apparel': 'Clothing',
    'Headgear': 'Clothing',
    'Lifestyle': 'Lifestyle Products',
    'Housewares': 'Household Items',
    'Waze': 'Waze Products',
    'Android': 'Android Devices',
    'Accessories': 'Accessories',
    'Fun': 'Lifestyle Products'
}

# Apply replacements in the 'Product_Category' column
data['Product_Category'] = data['Product_Category'].replace(category_mappings)

# Display the updated DataFrame
data.head()
```

```python
# Create a new column for total transaction amount
data['Total_Sales'] = data['Quantity'] * data['Average_Price']

# Display the updated DataFrame
data.head()
```

```python
# Create a new column for total cost to the consumer
data['Total_Cost'] = (data['Quantity'] * data['Average_Price']) * (1 - data['Discount_Percentage'] / 100) + data['Delivery_Charges']

# Display the updated DataFrame
data.head()
```

```python
# Create a new column for average dollars saved per transaction
data['Average_Saved($)'] = data['Average_Price'] * (data['Discount_Percentage'] / 100)

# Display the updated DataFrame
data.head()
```

```python
file_path = 'data/cleaned.csv'
data.to_csv(file_path, index=False)
```
